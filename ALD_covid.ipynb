```python
from taxcalc import *
```

```python
recs = Records()
pol = Policy()
calc_base = Calculator(pol, recs)
calc_base.advance_to_year(2020)
calc_base.calc_all()
```

```python
ref = {"ALD_Covid_c": {2020: [10000, 20000, 10000, 15000, 10000]},
       "ALD_Covid_ps": {2020: 78600},
       "ALD_Covid_prt": {2020: 1.0},
       "ALD_Covid_hc": {2020: 0}}
pol_ref = Policy()
pol_ref.implement_reform(ref)
calc_ref = Calculator(pol_ref, recs)
calc_ref.advance_to_year(2020)
calc_ref.calc_all()
```

```python
print("2020 Cost ($B):")
round((calc_base.weighted_total('combined') - calc_ref.weighted_total('combined')) / 1e9, 2)
```

```python
calc_base.difference_table(calc_ref, groupby='standard_income_bins', tax_to_diff='combined')
```
