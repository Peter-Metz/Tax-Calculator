```python
from taxcalc import *
```

```python
recs = Records()
pol = Policy()
calc_base = Calculator(pol, recs)
calc_base.advance_to_year(2020)
calc_base.calc_all()
```

# Scenario A: Above-the-line deduction

```python
refa1 = {"ALD_Covid_c": {2020: [10000, 20000, 10000, 15000, 10000]},
       "ALD_Covid_ps": {2020: 78600},
       "ALD_Covid_prt": {2020: 1.0},
       "ALD_Covid_hc": {2020: 0}}
pol_a1 = Policy()
pol_a1.implement_reform(refa1)
calc_a1 = Calculator(pol_a1, recs)
calc_a1.advance_to_year(2020)
calc_a1.calc_all()
cost_a1 = round((calc_base.weighted_total('combined') - calc_a1.weighted_total('combined')) / 1e9, 2)
print("Scenario A, with 100% phaseout")
print("2020 Cost ($B): {}".format(cost_a1))
calc_base.difference_table(calc_a1, groupby='standard_income_bins', tax_to_diff='combined')
```
